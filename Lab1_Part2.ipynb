```python
import requests
from bs4 import BeautifulSoup

URL = "https://books.toscrape.com/"

page = requests.get(URL)
print(page.status_code)

soup = BeautifulSoup(page.content, "html.parser")

results = soup.find_all("article", class_="product_pod")

print("Number of books:", len(results))
```

```python
first_book = results[0]
print(first_book)
```

```python
title_element = first_book.find("h3")
price_element = first_book.find("p", class_="price_color")
stock_element = first_book.find("p", class_="instock availability")

print(title_element)
print(price_element)
print(stock_element)
```

```python
print(title_element.text)
print(price_element.text)
print(stock_element.text)
```

```python
print(title_element.text.strip())
print(price_element.text.strip())
print(stock_element.text.strip())
```

```python
title = first_book.find("h3").find("a")["title"]
price = first_book.find("p", class_="price_color").text.strip()
stock = first_book.find("p", class_="instock availability").text.strip()

print(title)
print(price)
print(stock)
```

```python
print("\n***Book Element***")

for book in results:
    title = book.find("h3").find("a")["title"]
    price = book.find("p", class_="price_color").text.strip()
    stock = book.find("p", class_="instock availability").text.strip()

    print(title)
    print(price)
    print(stock)
    print()

print("***END***")
```

```python
print("\n***Available Books***")

available_books = []

for book in results:
    stock = book.find("p", class_="instock availability").text.strip()
    if "In stock" in stock:
        available_books.append(book)

print("Number of available books:", len(available_books))

for book in available_books[:10]:
    title = book.find("h3").find("a")["title"]
    price = book.find("p", class_="price_color").text.strip()
    stock = book.find("p", class_="instock availability").text.strip()

    print(title)
    print(price)
    print(stock)
    print()

print("***END***")
```
